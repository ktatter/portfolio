```python
columns_to_keep = ['AtomicNumber', 'Symbol', 'boiling_point', 'melting_point', 'density', 'is_radioactive', 'abundance_crust', 'abundance_sea']
df4 = df3[columns_to_keep].copy()
df4
```

```python
df4['abundance_crust'] = df4['abundance_crust'].fillna(0)
df4['abundance_sea'] = df4['abundance_sea'].fillna(0)
df4
```

```python
df4['perc_abundance_crust'] = 100 * df4['abundance_crust'] / df4['abundance_crust'].sum()
df4['perc_abundance_sea'] = 100* df4['abundance_sea'] / df4['abundance_sea'].sum()
df4
```

```python
cols_to_compare = ['Symbol','perc_abundance_crust','perc_abundance_sea']
df5 = df4[cols_to_compare].copy()
df5.sample(5)
```

```python
df5 = df5[(df5['perc_abundance_crust'] != 0) | (df4['perc_abundance_sea'] != 0)]
df5.head(15)
```

```python

# Set the theme
sns.set_theme(style="whitegrid")

# Reshape the DataFrame for better visualization with catplot
df5_melted = df5.melt(id_vars="Symbol", value_vars=["perc_abundance_crust", "perc_abundance_sea"],
                      var_name="Source", value_name="Abundance (%)")

# Create a catplot with 'bar' kind
g = sns.catplot(
    data=df5_melted, kind="bar",
    x="Symbol", y="Abundance (%)",
    hue="Source", palette=["green", "blue"],
    alpha=0.8, height=14
)

# Customize the plot
g.set_axis_labels("Symbol", "Abundance (%)")
plt.title("Abundance of Elements in Crust and Sea")

# Modify legend labels (from ChatGPT)
new_labels = ["Crust", "Sea"]
for t, l in zip(g._legend.texts, new_labels):
    t.set_text(l)

g.set(xlim=(-0.5, 54.5))
g.set(ylim=(0,100))

# Show the plot
plt.show();
```

```python
cols_for_comparison = ['AtomicNumber', 'Element', 'Symbol', 'AtomicMass', 'NumberofNeutrons', 'NumberofProtons', 'NumberofElectrons', 'Period', 'Group', 'Phase', 'Type', 'AtomicRadius', 'Electronegativity', 'FirstIonization', 'Density', 'MeltingPoint', 'BoilingPoint', 'SpecificHeat']
df_comparison = df3[cols_for_comparison].copy()
condition_1 = (df_comparison['Group'] >= 1) & (df_comparison['Group'] <= 2)
condition_2 = df_comparison['Group'] >= 13

# Applying the conditions
filtered_df_comparison = df_comparison[condition_1 | condition_2]
```

```python
filtered_df_comparison['Group'] = filtered_df_comparison['Group'].fillna(0).astype(int)
filtered_df_comparison.head()
```

```python
df_comparison = df_comparison.sort_values(by=['Period', 'Group'])
```

```python
# Set up the plot
plt.figure(figsize=(10, 6))

# Loop through each unique period and plot a line with a different color for each
for period, data in df_comparison.groupby('Period'):
    plt.plot(data['Group'], data['AtomicRadius'], label=f'Period {period}', marker='.')

# Add labels and title
plt.xlabel('Group')
plt.xticks(range(1, 19, 1))
plt.xlim(0.75,18.25)
plt.ylabel('Atomic Radius')
plt.yticks(np.arange(0, 3.5, 0.5))
plt.title('Atomic Radius vs Group for each Period')

plt.grid(True, which='major', linestyle='-', linewidth='0.1', color='black', alpha = 0.5)
plt.minorticks_off()  # Turn off minor ticks

# Add legend
plt.legend()

# Show the plot
plt.show();
```

```python
df3b = df3.copy()
location_mapping = {
    'UK/GB/Eng': ['United Kingdom', 'Great Britain', 'England', 'England/France', 'England/Sweden'],
    'Scotland': ['Scotland/Sweden'],
    'Germany': ['Sweden/Germany', 'Germany/France'],
    'France': ['England/France', 'Germany/France'],
    'Sweden': ['Sweden/Germany', 'Scotland/Sweden'],
    'Russia/USSR': ['Russia', 'USSR', 'USSR/United States', 'Russia/United States'],
    'United States': ['USSR/United States', 'Russia/United States']
    # Add more mappings as needed
}

# Map the original 'discovery_location' column using the dictionary
df3b['discovery_location'] = df3b['discovery_location'].map(lambda x: next((key for key, values in location_mapping.items() if x in values), x))

# Calculate the new counts after mapping
df_countries = df3b['discovery_location'].value_counts()

# Show the updated counts
print(df_countries)
```
